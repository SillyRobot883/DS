```python
x=np.arange(1,10)
y=np.arange(1,10)
```

```python
plt.figure(figsize=[20,5])
plt.suptitle('Subplots',fontsize = 12)


plt.subplot(1,2,1) 
plt.pie(np.arange(1,5), labels=['Jan','Feb','Oct', "Sep"], explode = [0.3,0,0,0])
plt.title('Months', fontsize = 15, )


plt.subplot(1,2,2)
plt.legend(['200','400','500'])
plt.plot([1,3,6,3], 'r')
plt.plot([2,3,4,6], 'b')
plt.plot([9,2,4,3], 'k')

plt.subplot(2,1,3)
plt.scatter(x,y, marker='o', alpha=0.4, s=50)
plt.title("Sport Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.grid(axis='y')
```

```python
plt.figure(figsize=[20,10])
plt.suptitle('Subplots',fontsize = 16)

#first
plt.subplot(2,2,1) #(number of rows, number of columns, number of plot)
plt.pie(np.arange(1,5),labels=['Jan','Feb','Mar', 'Apr']);
plt.title('Months')

#second
plt.subplot(2,2,2)
plt.plot([1,2,0,4,5])
plt.plot([10,9,8,6,3])
plt.legend(['2008','2009'], shadow = True, loc = 0)
plt.title('Years');

plt.subplot(2,2,3)
plt.scatter(x,y, marker='o', alpha=0.4, s=50)
plt.title("Sport Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.grid(axis='y')
```
